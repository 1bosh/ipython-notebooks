# Data Cleanup

We'll now perform these cleaup tasks one after another and then review the resulting dataset to see if additional cleanup tasks become apparent.

## Trim whitespace from a String cells

```python
cws = df[df['Country'].apply(lambda x: x.startswith(' '))]
len(cws)
```

```python
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
cws = df[df['Country'].apply(lambda x: x.startswith(' '))]
len(cws)
```

## Split Pos into 2 columns

```python
df[['Pos Num', 'Pos Abbrev']] = df['Pos'].str.extract('^(\d+)(\w+)$')
df.head(5)
```

## Split DOB/Age into 2 columns

```python
df[['DOB', 'Age']] = df['DOB/Age'].str.extract('^\((\d+-\d+-\d+).*aged (\d+)')
df['DOB'] = pd.to_datetime(df['DOB'])
df.head()
```

```python
del df['Pos'], df['DOB/Age']
df.head()
```

## Eliminating Duplicates

This is the most involving task in the cleanup phase. 

* Find string combinations with an edit distance less than 5 for each of the string columns to be evaluated.

```python
df[df['Club'].map(lambda x: 'München' in x)]
```

```python
import itertools
import nltk
import re

names = list(df['Country'].value_counts().keys())
combs = itertools.permutations(names, 2)

nc = []
for c in combs:
    if nltk.metrics.edit_distance(c[0], c[1]) < 3:
        nc.append(c)
nc[:10]
```
