```python
import itertools
import json

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Transpose the data so event types go into columns and days into rows. Also remove `Event` from the column lables. 

```python
df = pd.read_json('json/event_counts_by_day.json').transpose()
df.columns = df.columns.map(lambda x: x.replace('Event', ''))
```

Total number of events

```python
event_sums = df.sum()
event_sums.sum()
```

```python
event_sums.sort()
event_sums.plot(kind='barh', figsize=(14, 10), title='Events by frequency in 2013')
```

```python
top_events = event_sums.index[::-1][:5]
df[top_events].plot(figsize=(14, 10), title='Top 5 events over time')
```

```python
df.index = df.index.to_datetime()

# 0 = monday
df['Weekday'] = df.index.weekday
grouped = df.groupby('Weekday').sum()

grouped['Total'] = grouped.apply(sum, axis=1)

cols = grouped.columns
lencols = len(cols)

fig, axes = plt.subplots(nrows=6, ncols=3)
fig.suptitle('Distribution of all event types per weekday', y=1.01, fontsize=14)
fig.set_figheight(20)
fig.set_figwidth(14)

labels = ('Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun')

for idx, coords in enumerate(itertools.product(range(6), range(3))):
    if idx < lencols:
        ax = axes[coords[0], coords[1]]
        grouped[cols[idx]].plot(ax=ax, kind='bar')
        ax.set_title(cols[idx])
        ax.set_xticklabels(labels, rotation=0)
        ax.set_xlabel('', visible=False)
fig.tight_layout()
```
