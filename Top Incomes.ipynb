```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('csv/top-incomes.csv', skiprows=1)#, parse_dates=['Year'])
#countries = ['United States', 'China', 'Japan', 'Germany', 'France']
countries = ['United States', 'Japan', 'France']
footer = 'Data: parisschoolofeconomics.eu - Author: Ramiro Gómez - ramiro.org'
```

Show most recent data

```python
df.sort(columns=['Year'], ascending=False).head(10).dropna(axis=1, how='all')
```

```python
df_top = df[['Country', 'Year', 'Top 1% income share']].dropna(axis=0, how='any')
df_top[df_top['Year'] == 2011].sort(columns=['Top 1% income share'], ascending=False)
```

```python
df_top_gdp = df[df['Country'].isin(countries)]
df_top_gdp = df_top_gdp[['Year', 'Country', 'Top 1% income share']]
df_pivot = df_top_gdp.pivot('Year', 'Country', 'Top 1% income share').dropna(axis=0, how='all')

fig, axes = plt.subplots(nrows=1, ncols=1)
fig.set_figwidth(16)
fig.set_figheight(8)

df_pivot.plot(ax=axes, ylim=(0), xlim=(min(df_pivot.index), max(df_pivot.index)), linewidth=2)
axes.set_title('Evolution of the "top 1% income share" in France, Japan, and the United States')
axes.set_ylabel('Share in percent')
axes.set_xlabel('', visible=False)

fig.text(0, 0, footer, fontsize=12)
fig.tight_layout()
plt.show()
```
