```python
#url = 'http://www.kicker.de/news/fussball/bundesliga/vereine/1-bundesliga/2013-14/bayern-muenchen-14/trainer.html'
#url = 'http://www.kicker.de/news/fussball/bundesliga/vereine/1-bundesliga/2013-14/hamburger-sv-12/trainer.html'
url = 'http://www.kicker.de/news/fussball/bundesliga/vereine/1-bundesliga/2013-14/werder-bremen-4/trainer.html'
html = requests.get(url).text
soup = BeautifulSoup(html, 'html5lib')
```

```python
container = soup.find(id='slidercontainer')
```

```python
person_data = container.find_all(class_='daten')
match_data = container.find_all(class_='trainerbilanz')
```

```python
managers = defaultdict(dict)
for idx in range(len(person_data)):
    tds = person_data[idx].find_all('td')
    name = '%s %s' % (tds[1].text, tds[3].text)
    first_match = tds[12].text
    last_match = tds[14].text # 01.01.0001 for current manager
    period = {'first_match': first_match, 'last_match': last_match}

    managers[name]['name'] = name
    managers[name]['periods'] = managers[name].get('periods', []) + [period]
    
    rows = match_data[idx].find_all('tr')
    for row in rows[1:]:
        tds = row.find_all('td')
        # skip separator tables
        if 1 == len(tds):
            continue
        competition = tds[0].text
        matches = int(tds[1].text)
        win = int(tds[2].text)
        draw = int(tds[3].text)
        loss = int(tds[4].text)
        
        if 'competitions' not in managers[name]:
            managers[name]['competitions'] = {}

        if competition not in managers[name]['competitions']:
            managers[name]['competitions'][competition] = {
                'name': competition, 'matches': 0, 'win': 0, 'draw': 0, 'loss': 0}

        managers[name]['competitions'][competition]['matches'] += matches
        managers[name]['competitions'][competition]['win'] += win
        managers[name]['competitions'][competition]['draw'] += draw
        managers[name]['competitions'][competition]['loss'] += loss
```

```python
for name, data in managers.items():
    # missing data
    if 'competitions' not in data:
        continue
        
    data['totals'] = {'matches': 0, 'win': 0, 'draw': 0, 'loss': 0}
    for cname, competition in data['competitions'].items():
        data['totals']['matches'] += competition['matches']
        data['totals']['win'] += competition['win']
        data['totals']['draw'] += competition['draw']
        data['totals']['loss'] += competition['loss']
    data['totals']['points'] = data['totals']['win'] * 3 + data['totals']['draw']
    data['totals']['mean_points'] = round(data['totals']['points'] / data['totals']['matches'], 2)
```

```python
labels = []
values = []

for n, m in managers.items():
    if 'totals' not in m:
        continue
    labels.append('%s (%d)' % (n, m['totals']['matches']))
    values.append(m['totals']['mean_points'])

paired_sorted = sorted(zip(labels, values), key = lambda x: (x[1], -x[1]))
labels, values = zip(*paired_sorted)
    
graphs3.barh(labels, 
            values,
            'img/werder-bremen-managers-mean-points.de.png',
            figsize=(10, len(labels) * .5),
            title='Punktedurchschnitt pro Spiel über alle Wettbewerbe der Trainer vom Hamburger SV',
            footer='Daten: kicker.de | Quelle: torlaune.de | Erstellt: March 16, 2014')
```
