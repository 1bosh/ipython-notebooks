```python
import requests
import io
import re

import pandas as pd
import matplotlib.pyplot as plt

from datetime import datetime

url = 'http://wikitables.geeksta.net/dl/?url=https%3A%2F%2Fen.wikipedia.org%2Fwiki%2FMobile_operating_system&idx=0'

re_num = re.compile(r'^[\d,.]+$')

quarter_mapping = {'Q1': (3, 31), 'Q2': (6, 30), 'Q3': (9, 30), 'Q4': (12, 31) }


def fix_num(x):
    if (isinstance(x, str) and re.search(re_num, x)):
        return float(x.replace(',', ''))
    return x


def quarter_to_date(s):
    year, quarter = s.split(' ')
    dt = datetime(int(year), *quarter_mapping[quarter])
    return dt
```

```python
csv = requests.get(url).text
```

```python
df = pd.read_csv(io.StringIO(csv))
df = df.applymap(fix_num)
df['Date'] = df['Quarter'].apply(quarter_to_date)

df.set_index('Date', inplace=True)
df = df.sort().fillna(0) # sort by date and replace NaN with 0

df.to_csv('csv/mobile-os-market-share.csv')

del df['Total Phones']
del df['Total Smartphones']
del df['Quarter']

df.columns
```
