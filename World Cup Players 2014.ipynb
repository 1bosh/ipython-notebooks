# World Cup Players 2014

Some analysis and visualizations about the players set to participate in the football world cup 2014 based on [data from the Guardian](http://www.theguardian.com/football/datablog/2014/jun/06/world-cup-squads-rosters-broken-down-club-age-height).

See http://matplotlib.org/users/legend_guide.html#legend-guide

```python
import itertools

import pandas as pd
import matplotlib.pyplot as plt

from utils import graphs3

df = pd.read_csv('csv/world-cup-2014-players-guardian.csv')
footer = 'Data: theguardian - http://gu.com/p/3pzqy | CC BY-SA 2014 Ramiro Gómez - ramiro.org'
```

```python
df.head(5)
```

```python
del df['Player id']
```

## Players in domestic vs foreign leagues by country

Add Count and Domestic columns for doing calculations based in these figures.

```python
df['Count'] = 1
df['Domestic'] = df['Plays in home country?'].astype(int)
```

Group players by country and check whether there are 32 records.

```python
df_countries = df.groupby('Country')
assert 32 == len(df_countries)
```

```python
df_country_sums = df_countries.sum()
df_country_sums.head(5)
```

```python
df_domestic = df_country_sums[['Domestic', 'Count']].sort('Domestic', ascending=False)

ncols = 8
nrows = 4

fig, axes = plt.subplots(nrows=nrows, ncols=ncols)
fig.suptitle('FIFA World Cup 2014 Players in Domestic and Foreign Leagues', y=1.03, fontsize=20)
fig.set_figwidth(20)
fig.set_figheight(11)

for idx, coords in enumerate(itertools.product(range(nrows), range(ncols))):
    ax = axes[coords[0], coords[1]]
    row = df_domestic.ix[idx]
    vals = [row['Domestic'], row['Count'] - row['Domestic']]
    ax.set_title(df_domestic.index[idx])
    ax.pie(vals, labels=vals, colors=('b', 'r'))
    ax.set_xlabel('', visible=False)

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, ['Domestic League', 'Foreign League'], loc='lower right', bbox_to_anchor=(1, -0.26), fontsize=14)

text = 'Each pie chart shows the number of players for each participating country who play in domestic (blue) and foreign (red) leagues. Countries are sorted by the share of domestic players.\n{0}'.format(footer)
fig.text(0, -.02, text, fontsize=14)
fig.tight_layout()

plt.savefig('img/fifa-world-cup-2014-players-domestic-vs-foreign-leagues.png', bbox_inches='tight')
```

```python
df_goals = df_country_sums[['International goals']].sort('International goals')[-20:]
count = len(df_goals)
info = 'The top {0} countries by total number of international goals of their players.\n{1}'.format(count, footer)

graphs3.barh(df_goals.index, 
            df_goals['International goals'],
            'img/fifa-world-cup-2014-country-goals.png',
            figsize=(8, 8),
            title='FIFA World Cup 2014 Top {0} Countries by Player Goals'.format(count),
            footer=info)
```

## Top Clubs

```python
df_clubs = df['Club'].value_counts()[:30][::-1]
count = len(df_clubs)
info = 'The top {0} clubs by the number of world cup 2014 players who play there.\n{1}'.format(count, footer)

graphs3.barh(df_clubs.index, 
            df_clubs,
            'img/fifa-world-cup-2014-clubs.png',
            figsize=(8, 12),
            title='FIFA World Cup 2014 Top {0} Clubs'.format(count),
            footer=info)
```

## Top Club Countries

```python
df_club_countries = df['Club (country)'].value_counts()[:30][::-1]
count = len(df_club_countries)
info = 'The top {0} countries by the number of world cup 2014 players who play in clubs in these countries.\n{1}'.format(count, footer)

graphs3.barh(df_club_countries.index, 
            df_club_countries,
            'img/fifa-world-cup-2014-club-countries.png',
            figsize=(8, 12),
            title='FIFA World Cup 2014 Top {0} Club Countries'.format(count),
            footer=info)
```
