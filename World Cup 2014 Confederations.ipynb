```python
from io import StringIO
from utils import graphs3

import pandas as pd
import requests

r = requests.get('https://docs.google.com/feeds/download/spreadsheets/Export?key=1DOC8S4ydvgkAF59WwDFeAGqkAJsgjhHR-ATzFnlhknU&exportFormat=csv&gid=0')
data = r.text

df = pd.read_csv(StringIO(data), index_col=0)
```

```python
ro16 = df[['Qualified for WC', 'Round of 16']]

# Drop rows with all zeros
ro16 = ro16[(ro16 != 0).any(axis=1)]

# Calculate percentage
ro16['Percent in Round of 16'] = ro16['Round of 16'] / ro16['Qualified for WC'] * 100

ro16 = ro16.sort('Percent in Round of 16')
ro16
```

```python
title = 'Percentage of World Cup 2014 Teams by Confederation to Reach Knockout Stage'
footer = 'Data: fifa.com | Source: en.torlaune.de'
graphs3.barh(ro16.index,
             ro16['Percent in Round of 16'],
             'img/wc2014-percent-conferation-round-of-16.png',
             figsize=(10, 6),
             title=title,
             footer=footer)
```
