Some analysis and visualizations about the players set to participate in the football world cup 2014 based on [data from the Guardian](http://www.theguardian.com/football/datablog/2014/jun/06/world-cup-squads-rosters-broken-down-club-age-height).

See http://matplotlib.org/users/legend_guide.html#legend-guide

```python
import itertools

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('csv/world-cup-2014-players-guardian.csv')
```

```python
df.head(10)
```

```python
del df['Player id']
```

## Players in domestic vs foreign leagues by country

Add Count and Domestic columns for doing calculations based in these figures.

```python
df['Count'] = 1
df['Domestic'] = df['Plays in home country?'].astype(int)
```

Group players by country and check whether there are 32 records.

```python
df_countries = df.groupby('Country')
assert 32 == len(df_countries)
```

```python
df_country_sums = df_countries.sum()
df_country_sums.head(5)
```

```python
df_domestic = df_country_sums[['Domestic', 'Count']].sort('Domestic', ascending=False)

nrows = 6
ncols = 6
fig, axes = plt.subplots(nrows=nrows, ncols=ncols)
fig.suptitle('Share of FIFA World Cup 2014 Players in Domestic and Foreign Leagues', y=1.01, fontsize=18)
fig.set_figheight(14)
fig.set_figwidth(14)

for idx, coords in enumerate(itertools.product(range(nrows), range(ncols))):
    if idx >= 32:
        break
    row = df_domestic.ix[idx]
    vals = [row['Domestic'], row['Count'] - row['Domestic']]
    
    ax = axes[coords[0], coords[1]]
    ax.set_title(df_domestic.index[idx])
    ax.pie(vals, labels=vals, colors=('b', 'r'))
    ax.set_xlabel('', visible=False)

fig.tight_layout()
```
