```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import itertools

df = pd.read_csv('csv/top-incomes.csv', skiprows=1)#, parse_dates=['Year'])

gdp = ['United States', 'China', 'Japan', 'Germany', 'France']
g7 = ['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States']

countries = ['Australia', 'Canada', 'France', 'India', 'Japan', 'United States']

footer = 'Data source: parisschoolofeconomics.eu | Author: Ramiro Gómez - ramiro.org'
```

Show most recent data

```python
df.sort(columns=['Year'], ascending=False).head(10).dropna(axis=1, how='all')
```

```python
df_top = df[['Country', 'Year', 'Top 1% income share']].dropna(axis=0, how='any')
df_top[df_top['Year'] == 2011].sort(columns=['Top 1% income share'], ascending=False)
```

```python
dfs = df[df['Country'].isin(countries)]
dfs = dfs[['Year', 'Country', 'Top 1% income share']]
df_pivot = dfs.pivot('Year', 'Country', 'Top 1% income share').dropna(axis=0, how='all')

xmax = max(df_pivot.index)
xmin = xmax - 100

ncols = 2
nrows = int(len(countries) / ncols)

title = "Top 1% Income Share in Different Countries form {:d} to {:d}".format(xmin, xmax)

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, sharex=True, sharey=True)
fig.suptitle(title, y=1.03, fontsize=18)
fig.set_figwidth(12)
fig.set_figheight(10)

for idx, coords in enumerate(itertools.product(range(nrows), range(ncols))):
    ax = axes[coords[0], coords[1]]
    country = df_pivot.columns[idx]
    df_pivot[country].plot(
        ax=ax, 
        legend=False, 
        linewidth=2, 
        ylim=(0, 22), xlim=(xmin, xmax)
    )
    
    ax.set_title(country)
    ax.set_xlabel('', visible=False)

fig.text(.51, -0.01, footer, fontsize=12)
fig.tight_layout()
plt.show()
```
