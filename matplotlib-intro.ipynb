matplotlib is a library for making 2D plots of arrays in Python. matplotlib was originally influenced by MATLABÂ®, but can be used in a Pythonic, object oriented way. Internally matplotlib makes heavy use of NumPy and other extension code to provide good performance even for large arrays.

The matplotlib code is conceptually divided into three parts: the pylab interface for creating plots, the matplotlib API for creating and managing figures, text, lines, plots and the backends, which are device-dependent drawing devices, aka renderers, that transform the frontend representation to hardcopy or a display device.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
numbers = np.random.rand(100)
plt.plot(numbers)
plt.ylabel('random numbers')
plt.show()
```

```python
x = np.arange(0, 5, 0.1)
y = x**2
plt.plot(x, y, linewidth=4.0, color='#dddddd')
plt.ylabel('square')
plt.show()
```

```python
plt.gcf()
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)
plt.show()
```
