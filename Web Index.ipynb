```python
from lxml import html
import pandas as pd
import matplotlib.pyplot as plt
```

# Web Index Structure

http://thewebindex.org/data/all/structure/

```python
url ='http://thewebindex.org/data/all/scores/'
xpath = '//*[@id="tablepress-15"]'
 
tree = html.parse(url)
table = tree.xpath(xpath)[0]
raw_html = html.tostring(table)
 
df = pd.read_html(raw_html)[0]
df.to_csv('web-index-2013-all-scores.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(14, 10))
ax.scatter(df['Universal Access'], df['Education and awareness'], s=df['Web Index'], alpha=.5)
```
