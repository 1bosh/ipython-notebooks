Looking at the top 10 players by number of world cup tournaments a problem with the data becomes visible. Both Antonio Carbajal (Mexiko) (1950-1966) and Lothar Matthäus (1982-1998) [participated in 5 world cups](http://en.wikipedia.org/wiki/FIFA_World_Cup_records#Individual). Either the dataset is incomplete or something else went wrong.

Let's look at all players whose name contains `Lothar`.

```python
df[df['Player'].map(lambda x: 'Lothar' in x)]
```
