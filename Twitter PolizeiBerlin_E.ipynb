```python
# twitter archiver was run with Python 2.7 due to encoding issues: twitter-archiver --oauth PolizeiBerlin_E
import re
import pandas as pd

from utils import graphs
from __future__ import division

re_ht = re.compile(r'#(\S+)\b')
hashtags = {}
```

```python
with open('twitter/PolizeiBerlin_E', 'r') as f:
    tweets = f.read().splitlines()

tweet_count = len(tweets)
```

```python
for tweet in tweets:
    text = tweet.split('<PolizeiBerlin_E>')[1].strip()
    for tag in re.findall(re_ht, text):
        hashtags[tag] = hashtags.get(tag, 0) + 1
```

```python
df = pd.DataFrame(hashtags, index=[0]).transpose()
df.columns = ['Hashtag Count']
df['Hashtag Percent'] = df['Hashtag Count'] / tweet_count * 100
```

```python
df_top = df.sort('Hashtag Percent').tail(31)[:-1]

graphs.barh(df_top.index, 
            df_top['Hashtag Percent'].round(2),
            'img/Tweets-PolizeiBerlin_E-Top20-Hashtags.de.png',
            figsize=(10, 12),
            title='Anteil der Top 20 Hashtags in den %d Tweets von @PolizeiBerlin_E in Prozent ohne #24hPolizei' % tweet_count,
            footer='Daten: Tweets von @PolizeiBerlin_E | Autor: Ramiro Gómez - http://ramiro.org/')
```

```python
df_top = df.sort('Hashtag Percent').tail(31)[:-1]

graphs.barh(df_top.index, 
            df_top['Hashtag Percent'].round(2),
            'img/Tweets-PolizeiBerlin_E-Top20-Hashtags.en.png',
            figsize=(10, 12),
            title='Relative Frequency in %% of the top 20 hashtags in the %d Tweets by @PolizeiBerlin_E not counting #24hPolizei' % tweet_count,
            footer='Data: Tweets by @PolizeiBerlin_E | Author: Ramiro Gómez - http://ramiro.org/')
```
