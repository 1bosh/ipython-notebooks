I recently came across the [World Top Incomes Database](http://topincomes.g-mond.parisschoolofeconomics.eu/), which is compiled and maintained by Facundo Alvaredo, Tony Atkinson, Thomas Piketty and Emmanuel Saez. This is a quick dive into this dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt

import itertools

df = pd.read_csv('csv/top-incomes.csv', skiprows=1)#, parse_dates=['Year'])

# Manually selected countries with enough data
countries = ['Australia', 'Canada', 'France', 'India', 'Japan', 'United States']
footer = 'Data source: parisschoolofeconomics.eu | Author: Ramiro Gómez - ramiro.org'
```

## Determine the most recent year with data available

```python
df.sort(columns=['Year'], ascending=False).head(5).dropna(axis=1, how='all')
```

## Countries with Highest "Top 1% Income Share" in 2010

```python
df_top = df[['Country', 'Year', 'Top 1% income share']].dropna(axis=0, how='any')
df_top_2010 = df_top[df_top['Year'] == 2010].sort(columns=['Top 1% income share']).tail(10)
ax = df_top_2010.set_index('Country')['Top 1% income share'].plot(
    kind='barh',
    figsize=(12, 8), 
    title='Countries with Highest "Top 1% Income Share" in 2010')
ax.set_xlabel(footer)
ax.set_ylabel('')
plt.show()
```

## Evolution of "Top 1% Income Share" of Selected Countries

Show the evolution of the top 1% income share for several countries with fairly comprehensive data for the period from 1912 to 2012.

```python
dfs = df[df['Country'].isin(countries)]
dfs = dfs[['Year', 'Country', 'Top 1% income share']]
df_pivot = dfs.pivot('Year', 'Country', 'Top 1% income share').dropna(axis=0, how='all')

xmax = max(df_pivot.index)
xmin = xmax - 100

ncols = 2
nrows = int(len(countries) / ncols)

title = "Top 1% Income Share in Different Countries from {:d} to {:d}".format(xmin, xmax)

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, sharex=True, sharey=True)
fig.suptitle(title, y=1.03, fontsize=18)
fig.set_figwidth(12)
fig.set_figheight(10)

for idx, coords in enumerate(itertools.product(range(nrows), range(ncols))):
    ax = axes[coords[0], coords[1]]
    country = df_pivot.columns[idx]
    df_pivot[country].plot(
        ax=ax, 
        legend=False, 
        linewidth=2, 
        ylim=(0, 22), xlim=(xmin, xmax)
    )
    
    ax.set_title(country)
    ax.set_xlabel('', visible=False)

fig.text(.51, -0.01, footer, fontsize=12)
fig.tight_layout()

plt.savefig('img/evolution-top-1-percent-income-share.png', bbox_inches='tight')
```
