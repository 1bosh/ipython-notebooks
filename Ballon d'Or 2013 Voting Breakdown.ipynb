```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['axes.grid'] = False
plt.rcParams['grid.linewidth'] = 0
plt.rcParams['font.size'] = 13
plt.rcParams['axes.facecolor'] = 'ffffff'

df = pd.read_csv('csv/football/ballon-dor-male-players-votes-2013.csv')
```

```python
df.head()
```

```python
votes = []
vote_map = {'First (5 points)': 5, 'Second (3 points)': 3, 'Third (1 point)': 1}
for col, points in vote_map.items():
    df_new = df[['Vote', 'Country', 'Name', col]]
    df_new['Points'] = points
    df_new.columns = ['Group', 'Country', 'Voter', 'Voted', 'Points']
    votes.append(df_new)
df_votes = pd.concat(votes)
df_votes.head()
```

```python
total_points = df_votes['Points'].sum()
top_all = df_votes.groupby('Voted').sum().sort('Points').tail(20)
tally = top_all.Points.apply(lambda x: x / (total_points / 100.))
top_all
```
